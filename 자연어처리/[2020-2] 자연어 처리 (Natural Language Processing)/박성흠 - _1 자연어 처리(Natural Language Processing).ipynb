### 정규표현식

* 정규 표현식은 특정 문자들을 편리하게 지정하고 추가, 삭제 가능
* 데이터 전처리에서 정규 표현식을 많이 사용
* 파이썬에서는 정규 표현식을 지원하는 `re`(
regular express) 패키지 제공

* 정규 표현식 문법
  
| 특수문자 | 설명 |
| - | - |
| `.` | 앞의 문자 1개를 표현 |
| `?` | 문자 한개를 표현하나 존재할 수도, 존재하지 않을 수도 있음(0개 또는 1개) |
| `*` | 앞의 문자가 0개 이상 |
| `+` | 앞의 문자가 최소 1개 이상 |
| `^` | 뒤의 문자로 문자열이 시작 |
| `\$` | 앞의 문자로 문자열이 끝남 |
| `\{n\}` | `n`번만큼 반복 |
| `\{n1, n2\}` | `n1` 이상, `n2` 이하만큼 반복, n2를 지정하지 않으면 `n1` 이상만 반복 |
| `\[ abc \]` | 안에 문자들 중 한 개의 문자와 매치, a-z처럼 범위도 지정 가능 |
| `\[ ^a \]` | 해당 문자를 제외하고 매치 |
| `a\|b` | `a` 또는 `b`를 나타냄 |

* 정규 표현식에 자주 사용하는 역슬래시(\\)를 이용한 문자 규칙

| 문자 | 설명 |
| - | - |
| `\\` | 역슬래시 자체를 의미 |
| `\d` | 모든 숫자를 의미, [0-9]와 동일 |
| `\D` | 숫자를 제외한 모든 문자를 의미, [^0-9]와 동일 |
| `\s` | 공백을 의미, [ \t\n\r\f\v]와 동일|
| `\S` | 공백을 제외한 모든 문자를 의미, [^ \t\n\r\f\v]와 동일 |
| `\w` | 문자와 숫자를 의미, [a-zA-Z0-9]와 동일 |
| `\W` | 문자와 숫자를 제외한 다른 문자를 의미, [^a-zA-Z0-9]와 동일 |

#### match

* 컴파일한 정규 표현식을 이용해 문자열이 정규 표현식과 맞는지 검사

```
import re   # re 패키지를 불러올게
```

```
check ='ab.'    # .은 하나의 문자

print(re.match(check, 'abc'))   # 어떠한 문자든 .에 하나만 오면 되니까  
print(re.match(check, 'c'))     # ab.이랑 매칭 안됨
print(re.match(check, 'ab' ))   # 마찬가지로 .은 없어도 되는게 아니라 반드시 있어야함
```

#### search

* match와 다르게, search는 문자열의 전체를 검사

```
check = 'ab?'   # 앞의 b가 존재하거나 존재하지않아도 되는 문자열

print(re.search('a', check))    # a만 있어도 매칭 가능
print(re.match('kkkab', check)) # 앞의 k3개 니까 조건 안맞아. 시작이 k니까
print(re.search('kkkab', check))# 
print(re.match('ab', check))    # a다음 b가 올수도있고 안올수도있고

```

#### split

* 정규표현식에 해당하는 문자열을 기준으로 문자열을 나눔

```
r = re.compile(' ')                 # space bar를 compile하고 split을 한거지
print(r.split('abc abbc abcbab'))

r = re.compile('c')
print(r.split('abc abbc abcbab'))

r = re.compile('[1-9]')             # 1~9까지 split을 함
print(r.split('s1abc 2v3s 4sss 5a'))
```

#### sub

* 정규 표현식과 일치하는 부분을 다른 문자열로 교체


```
print(re.sub('[a-z]', 'abcdefg', '1'))  # pattern, replace, 1   # 이 문자열을 통제로 1로 바꿧네
print(re.sub('[^a-z]', 'abc defg', '1'))    # 이 문제열 제외하고 1로 바꿔줄게   # 매칭이 되지 않았으므로 그대로 나왓지
```

#### findall

* 컴파일한 정규 표현식을 이용해 정규 표현식과 맞는 모든 문자(열)을 리스트로 반환


```
print(re.findall('[\d]', '1ab 2cdd 3ef 4gh')) # 숫자에 대해서만 찾아줘

print(re.findall('[\W]', '!adbd@@#'))   # W (대문자) 특수문자 찾기 
# 일일이 찾아주는구만
```

#### finditer

* 컴파일한 정규 표현식을 이용해 정규 표현식과 맞는 모든 문자(열)을 `iterator` 객체로 반환
* `iterator` 객체를 이용하면 생성된 객체를 하나씩 자동으로 가져올 수 있어 처리가 간편함

```
iter1 = re.finditer('[\d]', '1ab 2cdd 3ef 4gh')
print( iter1)

for i in iter1:
    print(i)

print('\n','\n')
 
iter2 = re.finditer('[\W]', '!adbd@@#')
print(iter2)

for i in iter2:
    print(i)
```

### 토큰화(Tokenization)

* 특수문자에 대한 처리

  + 단어에 일반적으로 사용되는 알파벳, 숫자와는 다르게 특수문자는 별도의 처리가 필요            
  + 일괄적으로 단어의 특수문자를 제거하는 방법도 있지만 특수문자가 단어에 특별한 의미를 가질 때 이를 학습에 반영시키지 못할 수도 있음
  + 특수문자에 대한 일괄적인 제거보다는 데이터의 특성을 파악하고, 처리를 하는 것이 중요


* 특정 단어에 대한 토큰 분리 방법

  + 한 단어지만 토큰으로 분리할 때 판단되는 문자들로 이루어진 we're, United Kingdom 등의 단어는 어떻게 분리해야 할지 선택이 필요   
  + we're은 한 단어이나 분리해도 단어의 의미에 별 영향을 끼치진 않지만 United Kingdom은 두 단어가 모여 특정 의미를 가리켜 분리해선 안됨
  + 사용자가 단어의 특성을 고려해 토큰을 분리하는 것이 학습에 유리

#### 정규 표현식을 이용한 토큰화

* 토큰화 기능을 직접 구현할 수도 있지만 정규 표현식을 이용해 간단하게 구현할 수도 있음
* `nltk` 패키지는 정규 표현식을 사용하는 토큰화 도구인 `RegexpTokenizer`를 제공

```
from nltk.tokenize import RegexpTokenizer

sentence1 = 'Where there\'s a will, there\'s a way'
sentence2 = "Where there's a will, there's a way"

tokenizer = RegexpTokenizer('[\w]+')  # 소문자w, 문자와 숫자를 가져옴, 최소 1개 이상
tokens1 = tokenizer.tokenize(sentence1)
tokens2 = tokenizer.tokenize(sentence2)


print(tokens1)  # \랑 '를 제외함
print(tokens2)
```

```
tokenizer = RegexpTokenizer('[\s]+', gaps = True)       #소문자s, 정규표현식으로 지정
tokens1 = tokenizer.tokenize(sentence1)
tokens2 = tokenizer.tokenize(sentence2)
print(tokens1)
print(tokens2)
```

### n-gram 추출

* n-gram은 n개의 어절이나 음절을 연쇄적으로 분류해 그 빈도를 분석
* n=1일 때는 unigram, n=2일 때는 bigram, n=3일 때는 trigram으로 불림

```
from nltk import ngrams
sentence = 'There is no royal road to learning'
bigram = list(ngrams(sentence.split(), 2))  # 2 = bigram으로 해달라
bigram  # 2개끼리 하네
```

```
trigram = list(ngrams(sentence.split(), 3))
trigram

# 3개 끼리 중복되게
```

```
# textblob으로  2개 구분해볼게 bigram 처럼

from textblob import TextBlob

blob = TextBlob(sentence)
blob.ngrams(n=2)
```

```
blob.ngrams(n=3)
```

### 불용어 제거

* 영어의 전치사(on, in), 한국어의 조사(을, 를) 등은 분석에 필요하지 않은 경우가 많음
* 길이가 짧은 단어, 등장 빈도 수가 적은 단어들도 분석에 큰 영향을 주지 않음
* 일반적으로 사용되는 도구들은 해당 단어들을 제거해주지만 완벽하게 제거되지는 않음
* 사용자가 불용어 사전을 만들어 해당 단어들을 제거하는 것이 좋음
* 도구들이 걸러주지 않는 전치사, 조사 등을 불용어 사전을 만들어 불필요한 단어들을 제거


```
# 사용자가 불용어 사전을 만든다라..
# 뭐 금지어 인가 , 인위적인 조작이 가능하다는거네
# stop wards

stop_words = 'on in the'    # 일단 문장처럼 한개씩 썻네
stop_words = stop_words.split(' ')
stop_words
```

```
sentence = 'singer on the stage'
sentence = sentence.split(' ')


nouns = []          # 저장할 공간을 만들어주고

for noun in sentence:
    if noun not in stop_words:      # 만들어놓은 금지어목록에 없다면 append시켜버려
        nouns.append(noun)

nouns
```

### 정규표현식

* 한국어 정규 표현식도 대부분의 문법은 영어 정규 표현식과 같음
* 한국어는 자음과 모음이 분리되어 있기 때문에, 문법을 지정할 때는 자음과 모음을 동시에 고려해야 함

#### match

* 컴파일한 정규 표현식을 이용해 문자열이 정규 표현식과 맞는지 검사

```
import re

check = '[ㄱ-ㅎ]+'

print(re.match(check, 'ㅎ 안녕하세여')) #자음만 나오잖아

print(re.match(check, '안녕하세염 ㅎ')) # 처음부터 모음이 같이 있는 한글로 있기때문에 매칭 못함
```

#### search

* match와 다르게, search는 문자열의 전체를 검사

```
check = '[ㄱ-ㅎ|ㅏ-ㅣ]+'        # 모든 자음,모음에 대한 결과가 맞으면 return 하겠지

print(re.search(check, 'ㄱㅏ 안녕ㅋ'))
print(re.match(check, '안 ㄱㅏ'))   # match는 안되고
print(re.search(check, '안 ㄱㅏ'))   # search는 전체문장을 보기 때문에
```

#### sub

* 정규 표현식과 일치하는 부분을 다른 문자열로 교체

```
print(re.sub('[가-힣]','가나다라마바사', '1'))
print(re.sub('[^가-힣]', '가나다라마바사', '1'))    # ^ 아닌것만  1로 바꿔줘
```

## Bag of Words(BoW)

```
import re
from konlpy.tag import Mecab
tagger = Mecab()

corpus = '평생 살 것처럼 꿈을 꾸어라. 그리고 내일 죽을 것처럼 오늘을 살아라.'
tokens = tagger.morphs(re.sub('(\.)','',corpus))

vocab = {}
bow = []

for tok in tokens:
    if tok not in vocab.keys():
        vocab[tok] = len(vocab)
        bow.insert(len(vocab)-1,1)
    
    else:
        index = vocab.get(tok)
        bow[index] = bow[index]+1

print(bow)
print(vocab)
```

## 문서 단어 행렬(DTM)

* 문서 단어 행렬(Document-Term Matrix)은 문서에 등장하는 여러 단어들의 빈도를 행렬로 표현
* 각 문서에 대한 BoW를 하나의 행렬로 표현한 것

```
from sklearn.feature_extraction.text import CountVectorizer

corpus = ['Think like a man of action and act like man of thought.',
          'Try not to become a man of success but rather try to become a man of value.',
          'Give me liberty, of give me death']

vector = CountVectorizer(stop_words = 'english')
bow= vector.fit_transform(corpus)

print(bow.toarray())
print(vector.vocabulary_)

# 음 조금 불편하긴 하네
```

```
columns = []
for k, v in sorted(vector.vocabulary_.items(),key = lambda item:item[1]): # key values일단 다 받아오고 , dict이니까 // value기준으로 정렬하게 // 2번째꺼 가져오는거지
    columns.append(k)   # 뒤에 있는 index값으로 정렬할게 보기 편하게


import pandas as pd

df = pd.DataFrame(bow.toarray(), columns = columns)
df
```

## 어휘 빈도-문서 역빈도(TF-IDF) 분석

* 어휘 빈도-문서 역빈도(TF-IDF; Term Frequency-Inverse Docunment Frequency)는 단순히 빈도수가 높은 단어가 핵심어가 아닌, 특정 문서에서만 집중적으로 등장할 때 해당 단어가 문서의 주제를 잘 담고 있는 핵심어라고 가정

* 특정 문서에서 특정단어가 많이 등장하고 그 단어가 다른 문서에서 적게 등장할 때, 그 단어를 특정 문서의 핵심어로 간주
* 어휘 빈도-문서 역빈도는 어휘 빈도와 역문서 빈도를 곱해 계산 가능

* **어휘 빈도**는 특정 문서에서 특정 단어가 많이 등장하는 것을 의미

$$ tf_{x,y} $$

* **역문서 빈도**는 다른 문서에서 등장하지 않는 단어 빈도를 의미

$$ log(N/df_x) $$      

* **어휘 빈도-문서 역빈도**는 다음과 같이 표현

$$ W_{x,y} = tf_{x,y} * log(N/df_x) $$

* tf-idf를 편리하게 계산하기 위해 `scikit-learn`의 `tfidfvectorizer`를 이용
* 앞서 계산한 단어 빈도 수를 입력하여 tf-idf로 변환

```
# 바로 위에 했떤게 term frequency
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words = 'english').fit(corpus)

print(tfidf.transform(corpus).toarray())    # 계산된 tfidf 결과를 볼수가 있지
print(tfidf.vocabulary_)
```

* 좀 더 편리하게 확인하기 위해 데이터프레임으로 변환

```
# 백터값으로보자
# 이전에는 등장횟수엿지만
# 이제는 tfidf의 계산으로 나온거지
# 해당문서의 핵심어가 0.3 정도 나왔다
# 오히려 like가 0.6정도로 act 보다 횟수가 적지만 높은 값을 가지게 되었지


columns = []
for k, v in sorted(tfidf.vocabulary_.items(),key = lambda item:item[1]): # key values일단 다 받아오고 , dict이니까 // value기준으로 정렬하게 // 2번째꺼 가져오는거지
    columns.append(k)   # 뒤에 있는 index값으로 정렬할게 보기 편하게


import pandas as pd

df = pd.DataFrame(tfidf.transform(corpus).toarray(), columns = columns)
df

```
