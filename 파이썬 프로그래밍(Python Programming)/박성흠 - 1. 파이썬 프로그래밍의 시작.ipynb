### 1. 다음 코드를 각각 실행하고 결과를 보이시오.

`>>>` print("123" + "456")

###2. print() 함수를 사용하여 다음과 같이 출력하는 코드를 작성하시오.

-------
3+8*4-6/3 은 [결과값] 입니다.

*hint. print() 함수로 여러 개의 값을 출력할 때는 쉼표로 값들을 분리한다.

```
print(3+8*4-6/3,'은 [결과값]입니다')
print('3+8*4-6/3','은 [결과값]입니다')
```
