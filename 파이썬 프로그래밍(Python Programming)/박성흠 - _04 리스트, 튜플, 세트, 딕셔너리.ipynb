## 리스트(List)



```
print([])
#빈 리스트
print([1,2,3])
#리스트
print(['one','two','three'])
print([1,'one',2,'two'])
print([1,2,3,['onn','two']])
```

### 리스트 인덱싱(List Indexing)

* 리스트의 각 위치에 해당하는 값에 접근하기 위해서 주소 개념의 숫자를 사용

```
list = [1,2,3,4]
print(list)
print(list[1])
print(list[2])
print(list[3])
print([-1])
```

### 중첩 리스트 인덱싱(Nested List Indexing)

* 리스트 안에 리스트가 있을 경우 인덱스 접근 방법

```
list = [1,2,3,['one','tow']]
print(list)
print(list[1])
print(list[3][1])
```

### 리스트 슬라이싱(List Slicing)

* 리스트의 인덱스를 통해서 부분만 가져올 때 사용

```
list = ['one','two','three']
print(list)
print(list[0:])
print(list[1:2])
```

### 중첩 리스트 슬라이싱(Nested List Slicing)

* 중첩된 리스트에서 일부분만 인덱스를 통해서 가져올 때 사용

```
list=[1,2,3,['one','two','three']]
print(list)
print(list[2:4])
print(list[3][:2])
```

### 리스트 연산자/함수(List Operators/Function)

* 더하기 + 연산자
* 곱하기 * 연산자
* 리스트 길이를 구하는 `len()` 함수

```
list_1 = ['one','two','three']
list_2 = ['four','five','six']
print(list_1)
print(list_1+list_2)
print(len(list_1*3))
```

### 리스트 수정(List Modify)

* 리스트 인덱스 주소를 이용한 값 수정

```
list = ['one','two','three']
print(list)
list[2] = 3
print(list)
list[1]=2
print(list)
list[0]=1
print(list)
```

### 리스트 메소드(List Methods)


#### append()

* 리스트 요소 추가

```
list = ['one','two','three']
list.append('four')
print(list)
list.append([1,2,3,4])
print(list)
```

#### sort()

* 리스트 정렬

```
list = [1,4,2,5,3]
list.sort()
print(list)
list2=['orange','apple','banana','strawberry']
list2.sort()
print(list2)
```

#### reverse()

* 리스트 요소 반전

```
print(list)
list.reverse()
print(list)
```

#### index()

* 리스트의 요소 값에 대한 인덱스를 반환

```
list = [10,20,30,40,'g']
print(list)
print(list.index(10))
print(list.find('g'))
# list3 = 'abcedfg'
# list3.find('b')
```

#### insert()

* 리스트 요소 삽입

```
list = [10,20,30,40]
print(list)
list.insert(4,50)
print(list)
```

#### remove()

* 리스트 요소 제거

```
list = [10,20,30,40]
print(list)
list.remove(30)
print(list)
list.remove(40)
print(list)
```

#### del

* 리스트 요소 제거 연산

```
list = [10,20,30,40]
print(list)
del list[0]
print(list)
del list[0]
print(list)
```

#### pop()

* 리스트 요소를 방출

```
list = [10,20,30,40]
print(list)
list.pop()
print(list)
list.pop(0)
print(list)
```

#### count()

* 리스트 요소의 갯수 계산

```
list = [10,20,30,30,30]
print(list.count(30))
print(list.count(20))
```

#### extend()

* 리스트 확장

```
list_1 = [10,40,20,30]
print(list_1)
list_1.extend([50,60])
print(list_1)
list = [10,30,20,40]
list2 = [50,60]
print(list+list2)
```

## 튜플(Tuple)

* 리스트와 유사하지만 변경 불가능한 시퀀스 자료형
* 하나의 변수에 여러 값 할당 가능
* ‘(‘와 ‘)’를 사용하여 표현

```
print(())
print((1,2,3))
print(('one','two','three'))
print((1,2,3,('one','two')))
```

### 튜플 인덱싱(Tuple Indexing)

* 튜플의 각 위치에 해당하는 값에 접근하기 위해서 주소 개념의 숫자를 사용

```
tuple = ('one','two','three')
print(tuple)
print(tuple[1])
```

### 중첩 튜플 인덱싱(Nested Tuple Indexing)

* 튜플 안에 튜플이 중첩되어 있을 경우, 인덱스 접근 방법

```
t = (1,2,3,('one','two'))
print(t)
print(t[3])
print(t[3][1])
```

### 튜플 슬라이싱(Tuple Slicing)

* 튜플의 인덱스를 통해서 부분만 가져올 때 사용

```
t =(1,2,3)
print(t)
print(t[0:1])
print(t[1:])
```

### 중첩 튜플 슬라이싱(Nested Tuple Slicing)

* 중첩된 튜플에서 일부분만 인덱스를 통해서 가져올 때 사용

```
tuple = (1,2,3,('one','two','three'))
print(tuple)
```

### 튜플 연산자/함수(Tuple Operators/Function)

* 더하기 + 연산
* 곱하기 * 연산
* 튜플 길이를 구하는 `len()` 함수



```
tuple_1 = ('one','two','three')
tuple_2 = ('four','five','sex')
print(tuple_1)
print(tuple_2)
print(tuple_1+tuple_2)
print(len(tuple_1 + tuple_2))
```

## 세트(Set)

* 데이터 중복을 허용하지 않는 구조
* 순서가 없는 데이터 집합을 위한 구조
* 인덱싱으로 값을 접근할 수 없음

```
#집합
print({10,20,30,40})
#순서의 의미가 없다
# 어디다 쓰는 것인고하니
```

### 세트 연산자(Set Operators)

* 교집합: &
* 합집합: |
* 차집합: -
* 여집합: ^

```
#고유 기능이 있다
set1 = {1,2,3,4,5}
set2 = {4,5,6,6,7,8}
print(set1)
print(set2)
print(set1 & set2)
print(set1 | set2)
print(set1 - set2)
print(set1 ^ set2)
```

### 세트 메소드(Set Methods)

* 교집합: intersection()
* 합집합: union()
* 차집합: difference()
* 여집합: symmetric_difference()

```
set1 = {1,2,3,4,5}
set2 = {4,5,6,6,7,8}
print(set1)
print(set2)

#연산자 접근, 메서드 접근 가능

print(set1.intersection(set2))
print(set1.union(set2))
print(set1.symmetric_difference(set2))
```


* 요소 추가: add()
* 여러 요소 추가: update()
* 요소 제거: remove()
* 요소 제거: discard()
* 모든 요소 제거: clear()

```
set = {10,20,30,40}
print(set)
set.add(50)
print(set)
set.update([60,70])
print(set)
set.remove(70)
set.remove(60)
print(set)

set.discard(30)
print(set)

set.clear()
print(set)
```
