### 1. 세 개의 양의 정수를 입력받아 그 합이 짝수이면 가장 큰 수를 출력하고, 홀수이면 가장 작은 수를 출력하는 프로그램을 작성하시오.

```
def pro(*n):
    # list = [n]
    # list.sort()
    # if sum(list) % 2 == 0:
    #     print(list[0])
    # elif sum(list) % 2 == 1:
    #     print(list[-1])
    result = 0
    for i in n:
        result = result +i
    list = []                   # 튜플이니까 항쌍 for을 통해 
   
    for j in n:
        list.append(j)          #가변 매개변수를 쓰면 tuple로 들어가니까 list로 하나씩집어넣어주고
    list.sort()

    if result % 2 == 0:
        return list[-1]
    elif result % 2 == 1:
        return list[0]

pro(14,3,7)
# pro(1,2,8)  #홀수니까
```

### 2. 두 점 사이의 거리를 구하는 프로그램을 작성하시오.


*   두 점의 x, y 좌표를 각각 입력 받는다.
*   두 점 사이의 거리를 구하는 공식은 다음과 같다.
$${\sqrt{(x1-x2)^2 + (y1-y2)^2}}$$

*   제곱근은 (~)**0.5 로 표현한다.





```
def distance(x1,y1,x2,y2):
    return ((x1-x2)**2+(y1-y2)**2)**0.5

distance(0,0,1,1)
```

### 4. 사각형의 둘레를 구하는 함수를 작성하고, 가로 세로의 길이가 각각 3, 6인 사각형의 둘레를 함수를 이용하여 구해 보시오.

```

# Rectangle circumference
def rectangle_circumference(a,b):
    return 2*a+2*b
rectangle_circumference(3,6)
```

### 5. 리스트로 작성한 학생 점수를 입력하면 최고점을 구하는 함수 high_score를 정의하고, 이 함수를 이용하여 아래 점수 중 최고점을 구하는 프로그램을 작성하시오.

score = [60, 72, 98, 48, 100, 86]

```
score = [60,72,98,48,100,86]
def high_score(n):
    n.sort()
    return n[-1]

high_score(score)
```
