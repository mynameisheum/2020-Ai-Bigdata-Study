### 파일 모드(File Mode)


### 텍스트 파일 쓰기


#### write()

* `write()`를 이용하여 파일에 쓰기

```
f = open('file.txt', 'w')
# 쓰기 모드로 file.txt열고
f.write('hello python')
# hello python 쓰고
f.close()
# 닫았찌
```

```
!cat file.txt
#cat 명령어로 저장
```

#### writelines()

* `writelines()`를 이용하여 list를 파일에 쓰기

```
list = ['one\n','two\n','three\n']
#\n으로 구분할수있음을..

f = open('file.txt','w')
f.writelines(list)
f.close()
```

```
!cat file.txt
```

#### 표준 입력 → 파일 쓰기

* 표준 입력을 `input()` 함수를 통해 입력 받고, `write()` 함수를 통해 파일에 쓰기

```
text = input('입력 : ')
f = open('file.txt', 'w')
f.write(text)
f.close()
```

```
!cat file.txt
```

```
#여러가지를 넣고 싶을떈?
text = [str(text + '\n')for text in input('여러 값 입력 : ').split()]
# print(text)
f = open('file.txt','w')
f.writelines(text)
f.close()
```

```
!cat file.txt
```

```
x = ''
text = ''
while x != 'q':
    x = input('반복입력 : ')
    text += x + '\n'

f = open('file.txt','w')
f.writelines(text)
f.close()
```

```
!cat file.txt                                                                                                                                                     
```

### [Lab] 구구단 결과를 파일에 쓰기

* 구구단 결과를 파일에 쓰기

```
f = open('dan.txt','w')
for i in range(2,10):
    for j in range(1,10):
        f.write('{0} x {1} = {2}\n'.format(i,j,i*j))
    else:
        f.write('\n')
f.close()
```

#### readline()

* `readline()`를 이용하여 텍스트 파일을 라인 단위로 읽고 화면에 출력

```
f = open('anthem.txt', 'r')
while True:
    line = f.readline()
    if not line :break
    print(line)

f.close()
```

#### readlines()

* `readlines()`를 이용하여 텍스트 파일의 여러 라인을 읽고 화면에 출력

```
f = open('anthem.txt','r')
lines = f.readlines()
print(lines)
f.close()
```

#### read()

* `read()`를 이용하여 텍스트 파일을 읽고 화면에 출력

```
f = open('anthem.txt','r')
data = f.read()
print(data)
f.close()
```

#### tell()

* 파일 포인터 위치 변경

```
f = open('anthem.txt','r')
while True:
    print(f.tell())         #file의 point의 위치 알려주기
    line = f.readline()
    if not line:
        break
    print(line)

f.close()
```

#### seek()

* 파일 포인터 위치 이동

```
f = open('anthem.txt','r')
f.seek(43)                      #46번째로 seek으로 이동
line = f.readline()
print(line)

f.seek(0)
line = f.readline()
print(line)
f.close()
```

### with 문

* 항상 파일을 `open()` 함수로 열고, `close()` 함수로 닫아야 하는 일을 자동으로 처리
* with문을 이용하면 with 블록 내에서 파일을 열고 벗어나면 파일을 닫음

```
f = open('file.txt','w')
f.write('hello python')
f.close()
```

```
with open('file.txt','w') as f:
    f.write('hi there')
```

* 사용자로부터 표준 입력 받음

### 디렉토리 및 파일 확인

* `isdir()`: 디렉토리 존재 확인
* `isfile()`: 파일 존재 확인
* `exists()`: 디렉토리/파일 존재 확인

```
import os.path
print(os.path.isdir('test'))
print(os.path.isdir('test/1'))
print(os.path.isdir('test/4'))  # 4는 없으니까 false
print(os.path.isdir('test2'))
print(os.path.isdir('test2/1'))
print('----------------------')
print(os.path.isfile('unname.jpg'))
print(os.path.isfile('KakaoTalk__20200713_234421669_14.jpg'))
print('----------------------')
print(os.path.isdir('test'))
print(os.path.isdir('test/1'))
print(os.path.exists('test/4'))  # 4는 없으니까 false
print(os.path.exists('test2'))
print(os.path.exists('test2/1'))
```

### 파일 압축

1. zipfile:압축 관련 라이브러리
2.ZipFile() : 압축 파일 열기
3. write() : 압축 파일에 쓰기
4. extractall() : 전체 압축 해제
5. close() : 압축 파일 닫기


```
import zipfile
comp = zipfile.ZipFile('new.zip','w')
comp.write('KakaoTalk_20200713_234421669_14.jpg')
comp.close()
```

```
!ls
```

```
decomp = zipfile.ZipFile('new.zip','r')
decomp.extractall('new')    # 압축 풀 디렉토리를 정해주는 거
decomp.close()
```
