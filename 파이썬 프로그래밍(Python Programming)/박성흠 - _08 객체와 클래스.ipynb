### Book 클래스 메소드 정의

* 메소드
  + 책 정보 출력: `print_info(self)`
  + `self`가 있어야만 실제로 인스턴스가 사용할 수 있는 메소드로 선언

```
class Book(object):
    author = ''
    title = ''
    publisher = ''
    data = ''

    def print_info(self):           # 자기 자신의 인스턴스를 self로 불러옴
        print('author : ', self.author)
        print('title : ', self.author)
```

```
book = Book()       #Book이라는 class를 통해 book이라는 인스턴스를 만들고

book.title = 'python'   #정해주고
book.author = 'heum'    #정해주고

book.print_info()   #print_info를 호출하고 (self)= book이라 생각하면 됨
```

## 클래스 매직 메소드(Class Masic Methods)


### `__init__()`

* `__init__()` 메소드를 이용하여 클래스의 속성들을 초기화

```
class Book(object):
    count = 0

    def __init__(self,author,title,publisher,data):
        self.author = author
        self.title = title
        self.publisher = publisher
        self.data = data
        Book.count += 1

    def print_info(self):           # 자기 자신의 인스턴스를 self로 불러옴
        print('author : ', self.author)
        print('title : ', self.title)
        print('publisher : ',self.publisher)
        print('date : ',self.data)
```

```
book = Book('heum','kiki','colap','2020')
book.print_info()
print('number of books',str(Book.count))
```

### `__str__()`

* `__str__()` 메소드를 이용하여 인스턴스 출력

```
class Book(object):
    count = 0

    def __init__(self,author,title,publisher,data):
        self.author = author
        self.title = title
        self.publisher = publisher
        self.data = data
        Book.count += 1

    def __str__(self):

        return('author : ' + self.author + \
               '\ntitle : ' + self.title + \
               '\npublisher : ' + self.publisher + \
               '\ndata : ' + self.data)
```

```
book = Book('heum','kiki','colap','2020')
print(book)
# book.print_info()
print('number of books',str(Book.count))
```

### 매직 메소드 예제

* Line 클래스

```
class Line(object):
    length = 0

    def __init__ (self, length):                #인스턴스의 속성값
        self.length= length                     #인스턴스의 속성값을 잘 초기화 시켜주고
        print(self.length, "길이의 선 생성")    # 그걸 호출한거지
    
    def __del__(self):
        print(self.length, '길이선 제거')

    def __repr__(self):
        return str(self.length)

    def __add__(self,other):
        return self.length + other.length

    def __lt__(self,other):
        return self.length < other.length

    def __le__(self,other):
        return self.length <= other.length
    
    def __gt__(self,other):
        return self.length > other.length
    
    def __ge__(self,other):
        return self.length >= other.length

    def __eq__(self ,other):
        return self.length == other.length

    def __ne__(self,other):
        return self.length != other.length
```

```
l1 = Line(10)
print(l1)

l2 = Line(20)
print(l2)

print('선의합 : ', l1 + l2)

if l1 < l2:
    print( l1, '<', l2)
elif l1 <= l2:
    print( l1, '<=', l2)
elif l1 > l2:
    print( l1, '>', l2)
elif l1 >= l2:
    print( l1, '>=', l2)
elif l1 == l2:
    print( l1, '==', l2)
elif l1 != l2:
    print(l1, '!=', l2)
else:
    pass
```

### 가시성 예제

* `__items` 속성은 Box 객체 외부에서 보이지 않도록 캡슐화와 정보 은닉이 가능
* 외부에서 `__items` 속성에 접근하면 속성 오류 발생

```
class Box(object):
    def __init__(self,name):
        self.name = name
        self.__items = []       #일단 초기화로 아무것도 없게 만들기

    def add_item(self,item):
        self.__items.append(item)
        print('아이템 추가')
    
    def get_number_of_items(self):
        return len(self.__items)    #갯수는 알수잇지만 아이템이 뭔지는 알수없음
```

```
box = Box('Box')
box.add_item('Item1')
box.add_item('Item2')
print(box.name)
print(box.get_number_of_items())

print(box.__items)              #은닉화,캡슐화
```

```
class SuperClass(object):
    pass
class SubClass(SuperClass):
    pass
```

### 메소드 오버라이딩(Method Overriding)




```
# 약간 변경해서 사용하고 싶다

class SuperClass(object):
    def method(self):
        pass

class SubClass1(SuperClass):
    def method(self):
        print('Method Overriding')

class SubClass2(SuperClass):
    pass
```

```
sub1 = SubClass1()
sub2 = SubClass2()

sub1.method()
sub2.method()       #SubClass2에는 method가 없지만 SubClass1을 상속 받았으니
                    # SubClass1안에 있는 method를 사용하게 된거지
```

### 클래스 상속, 메소드 오버라이딩 예제

* Vehicle 클래스를 상속받아 Car 클래스와 Truck 클래스 생성
* Car 클래스와 Truck 클래스는 up_speed 메소드를 오버라이딩
* Car 클래스는 속도가 240 초과되면 240으로 조정
* Truck 클래스는 속도가 180 초과되면 180으로 조정

```
class Vehicle(object):
    speed = 0
    def up_speed(self,value):
        self.speed += value
    
    def up_down(self,value):
        self.speed -= value

    def print_speed(self):
        print('speed',str(self.speed))

class Car(Vehicle):
    def up_speed(self, value):
        self.speed += value
        if self.speed > 200:
            self.speed = 240        #한계 정하기

class Truck(Vehicle):
    def up_speed(self, value):
        self.speed += value
        if self.speed > 180:
            self.speed = 180          
```

```
car = Car()
car.up_speed(300)
car.print_speed()

truck = Truck()
truck.up_speed(200)
truck.print_speed()
                    # 200 속도를 넣어줘도 180으로 limit를 거는거지
```
