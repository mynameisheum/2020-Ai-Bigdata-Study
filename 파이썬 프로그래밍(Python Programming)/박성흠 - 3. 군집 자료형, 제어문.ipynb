### 3. 리스트를 원소로 갖는 리스트의 각 원소들의 합을 구하는 프로그램을 작성하시오.

a = [(1,2), (3,4),(6,7),(10,11)]

(hint) for (n1,n2) in a:



```
a = [[1,2],[3,4],[6,7],[10,11]]
for n1,n2 in a:
    print(n1+n2)
```

### 5. 2020년이 윤년인지 검사하는 프로그램을 작성하시오.


*  400 으로 나누어 떨어지는 해는 윤년이다.
*  4로 나누어 떨어지는 해는 윤년이다.
*  4로 나누어 떨어지지만 100으로 나누어 떨어지는 해는 윤년이 아니다.




```
year = 2020
if year % 400 == 0 and year % 4 == 0:
    print(year,'은 윤년입니다')
elif year % 4 and year % 100 == 0:
    print(year,'은 윤년이 아닙니다')
else:
    print(year,'은 윤년이 아닙니다')
```
