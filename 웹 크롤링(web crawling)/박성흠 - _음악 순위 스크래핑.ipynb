```
# 2page까지 불러부자
# 함수 만들어서

from bs4 import BeautifulSoup
import urllib.request
import pandas as pd


def get_music_ranks(url):

    search_url = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(search_url,'html.parser')
    
    soup_titles = soup.select('a._title span.ellipsis') # class 니까 .으로 불러오기
    titles = [title.get_text() for title in soup_titles]


    soup_artists = soup.select('td._artist.artist > a')    # span까지 인식을 못하니까 빼자 
    artists = [artist.get_text().strip() for artist in soup_artists]
    
    print(len(titles))
    print(len(artists))
    return titles,artists

url_p1 = 'https://music.naver.com/listen/top100.nhn?domain=TOTAL_V2&page=1'
url_p2 = 'https://music.naver.com/listen/top100.nhn?domain=TOTAL_V2&page=2'

titles_50, artists_50 = get_music_ranks(url_p1)
title_50_100, artists_50_100 = get_music_ranks(url_p2)

ranks = [rank for rank in range(1,101)]
rank_df = pd.DataFrame({'rank' : ranks,
                        'title' : titles_50 + title_50_100,
                        'artist' : artists_50 + artists_50_100})

rank_df

# 디자인이나 코드를 바꾸지 않는 이상 계속 작동함
# 벗 매번 run 할때마다 새로운걸로 갱신 되겠지
```
