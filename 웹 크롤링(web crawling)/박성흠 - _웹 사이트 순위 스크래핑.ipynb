# BeautyfulSoup 웹 사이트 순위 스크래핑

* Alexa: https://www.alexa.com/topsites

## 글로벌웹사이트 순위

```
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd

url = 'https://www.alexa.com/topsites'  # 사이트 주소
search_url = urllib.request.urlopen(url).read()
soup = BeautifulSoup(search_url, 'html.parser')
rank_df = pd.DataFrame(columns = ('rank','site','daily time on site','Daily Pageviews per Visitor', '% of Traffic From Search', 'Total Sites Linking In'))

sites = soup.find_all('div', {'class':'tr site-listing'})   # tr이라는 부분만 읽고있으니까

for site in sites:  #row의 tr해당하는 거중에 td값을 하나씩 맵핑해줘야겟지
    rank = site.find('div',{'class':'td'}).get_text()
    website = site.select('p > a')[0].get_text()    # 리스트요소중에 첫번째만 올거라 예상하는거지
    infos = site.find_all('div', {'class': 'td right'}) # td right가 4개있잖아
    time = infos[0].get_text()
    pageview = infos[1].net_text
    traffic = infos[2].net_text
    linking = infos[3].net_text
    rank_df.loc[rank]  = [rank, website, time, pageview, traffic, linking]  # rank값을 index로 사용하는거지

rank_df
```
