```python
import numpy as np
```

```python
# 모두 0으로 채운 3 x 3 배열 a 생성 
a = np.zeros((3,3))
print(a)

# 모두 1로 채운 2 x 3 배열 b 생성
b = np.ones((2,3))
print(b)

# 5로 채운 2 x 3 x 3 배열 c 생성
c = np.ones((2,3,3))+np.ones((2,3,3))+np.ones((2,3,3))+np.ones((2,3,3))+np.ones((2,3,3))
print(c)

# 5 x 5 단위 행렬 d 생성
d = np.eye(5)
print(d)
```

```python
# range, reshape 함수를 사용하여 1~20 의 정수로 4 x 5 배열 arr 생성
g = np.arange(1,21).reshape(4,5)
print(g)
```

```python
# 불린 인덱스를 이용하여 arr 배열의 대각선 '\' 상의 요소 선택
m = [[True,False,False],
    [False,True,False],
    [False,False,True]]

m2 = [True,False,False]

arr = np.random.randint(1,20,(3,3))
# b1 = np.random.randint(0,2,(3,3),dtype = bool)
b2 = np.array([[1,0,0],[0,1,0],[0,0,1]], dtype = bool)
print(arr)
print(arr[b2])
# print(arr[m2])
# arr[m2]
# arr[m]
```

```python
# arr 배열의 인덱스가 짝수인 요소만 조회 (불린 인덱스와 조건식을 조합)
a2 = np.random.randint(1,10,10)
print(a2)
for i in range(5):
    a2[2*i] = True

t = []

for j in range(len(a2)):
    if a2[j] is True:
        append.t(a2[j])

t
```

# 배열 간 연산

```python
m1 = np.array([
    [1,2],
    [3,4]
])
m2 = np.array([
    [5,6],
    [7,8]
])

# m1, m2 배열의 각 요소 간 합, 차, 곱 그리고 행렬곱을 구하기
print(m1 + m2)
print(m1-m2)
print(m1*m2)
print(m1.dot(m2))
```

```python
# arr 배열의 컬럼(열)별 합계
# print(sum(m1[0]))
# print(sum(m1[1]))
print(m1)
print(np.sum(m1,axis = 0))
# print(np.sum(m1, axis = 1))
```

```python
# arr 배열의 행별 평균
# print(np.mean(m1,axis = 0))
print(m1)
print(np.mean(m1,axis =1))
```
