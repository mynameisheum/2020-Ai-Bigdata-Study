#### Index 연산

```
# append
# difference
# intersection
# union
# isin
# delete
# drop
# insert
# in_monotonic
# os_unique
# unique
```

```
idx1 = pd.Index([1,2,4,6,8])
idx2 = pd.Index([2,4,5,6,7])
print(idx1.append(idx2))

#append 사용 가능 
#list처럼 중복되는 것도 그냥 쭉 써줌
```

```
print(idx1.difference(idx2))
print(idx1 - idx2)
#차집합
#배열의 개념이야
```

```
print(idx1.intersection(idx2))
print(idx1 & idx2)
```

```
print(idx1.union(idx2))
print(idx1 | idx2)
# union은 중복 안된 형태로 
# 합집합이지
```

```
print(idx1.delete(0))

# 인덱스 위치로 빠지게 한거고
# 맨 처음 값이 빠짐

print(idx1.drop(6))

#지정된 값을 빼줌
```

```
print(idx1 ^ idx2)
#공통된 나머지를 뺌
```
